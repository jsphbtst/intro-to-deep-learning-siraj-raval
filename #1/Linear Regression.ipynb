```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
sns.set_style("whitegrid", {"axes.grid": True})
```

```python
df = pd.read_csv('challenge_dataset.txt')
df.head()
```

```python
def linear_regression(X, y, m_current=0, b_current=0, max_iterations=10000, tol=1e-5, learning_rate=0.0001):
    N = float(len(y))
    num_iterations = 0
    m_values = b_values = cost_values = []
    cost = 2
    while cost > tol:
        y_current = (m_current * X) + b_current
        cost = sum([data**2 for data in (y-y_current)]) / N
        
        m_gradient = -(2/N) * sum(X * (y - y_current))
        b_gradient = -(2/N) * sum(y - y_current)
        
        m_current = m_current - (learning_rate * m_gradient)
        b_current = b_current - (learning_rate * b_gradient)
        
        m_values.append(m_current)
        b_values.append(b_current)
        cost_values.append(cost)
        
        num_iterations = num_iterations + 1
        if num_iterations > max_iterations:
            break
    return m_current, b_current, cost, m_values, b_values, cost_values
```

```python
m_current, b_current, cost, m_values, b_values, cost_values = linear_regression(df['X'], df['y'])
```

```python
x_range = np.array(range(int(min(df['X'])), int(max(df['X']))))

plt.figure(figsize=(15,7))
plt.title('X vs y in challenge_dataset', fontsize=20)
plt.xlabel('X', fontsize=20)
plt.ylabel('y', fontsize=20)

plt.scatter(df['X'], df['y'], color='red', label='Data Points')
plt.plot(x_range, m_current*x_range + b_current, color='blue', label='Regression Line', linewidth=3)

plt.legend(loc='lower right', fontsize=25)
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(15,7))
ax = Axes3D(fig)

ax.scatter(m_values, b_values, cost_values)

ax.set_xlabel('SLOPE (M) VALUES', fontsize=15)
ax.set_ylabel('Y-INTERCEPT (B) VALUES', fontsize=15)
ax.set_zlabel('COST VALUES', fontsize=15)
```
